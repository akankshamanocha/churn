```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_excel('Case_study_data.xlsx')
```

```python
df
```

```python
service=pd.get_dummies(df['SERVICE_ PLAN'],drop_first=True)
model=pd.get_dummies(df['MODEL'],drop_first=True)
sales_ch=pd.get_dummies(df['SALES_CHANNEL'],drop_first=True)
direct_indirect=pd.get_dummies(df['DIRECT_INDIRECT_CHANNEL'],drop_first=True)
city=pd.get_dummies(df['CITY'],drop_first=True)
state=pd.get_dummies(df['STATE'],drop_first=True)
feedback=pd.get_dummies(df['FEEDBACK'],drop_first=True)
most_recent=pd.get_dummies(df['MOST_RECENT_SALES_CHANNEL'],drop_first=True)
site_type=pd.get_dummies(df['SITE_TYPE_DESC'],drop_first=True)
comp_owner=pd.get_dummies(df['COMPUTER_OWNER'],drop_first=True)
acc_status=pd.get_dummies(df['ACCOUNT_STATUS'],drop_first=True)
```

```python
churn=pd.concat([df,service,model,sales_ch,direct_indirect,city,state,feedback,most_recent,site_type,comp_owner,acc_status],axis=1)
```

```python
churn.columns
```

```python
churn.drop(['SITE_ACCOUNT_NUMBER', 'ACCOUNT_STATUS'],axis=1,inplace=True)
```

```python
churn.drop(['MODEL','SALES_SOURCE','FIRST_NAME','LAST_NAME','ADDRESS','CITY','STATE'],axis=1,inplace=True)
```

```python
churn.drop(['COMPANY_SOURCE_NAME','EMAIL_ADDRESS','FEEDBACK','SATELLITE','SITE_TYPE_DESC','WARRANTY_NAME','MOST_RECENT_SALES_CHANNEL'],axis=1,inplace=True)
```

```python
churn.drop(['DATE_OF_BIRTH'],axis=1,inplace=True)
```

```python
churn.drop(['COMPUTER_OWNER'],axis=1,inplace=True)
```

```python
churn.drop(['FIRST_COMMUNICATION_DATE','UPGRADE_DATE'],axis=1,inplace=True)
```

```python
churn.drop(['INCOME'],axis=1,inplace=True)
```

```python
churn.convert_objects(convert_numeric=True)
```

```python
churn.fillna(0,inplace=True)
```

```python
churn.drop(['Unnamed: 29'],axis=1,inplace=True)
```

```python
churn.drop(['SERVICE_ PLAN','SALES_CHANNEL'],axis=1,inplace=True)
```

```python
X=churn.drop('Closed',axis=1)
y=churn['Closed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
predictions=log.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
